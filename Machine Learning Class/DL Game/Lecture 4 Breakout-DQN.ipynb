```python
DISPLAY_H = 400
DISPLAY_W = 600
BG_IMG = 'background.jpg'

FPS = 30

PADDLE_START_X = 300
PADDLE_START_Y = 380
PADDLE_IMG = 'paddle.jpg'

BALL_RADIUS = 10
BALL_SPEED = 6
BALL_RECT = int(BALL_RADIUS * 2)
DX, DY = 1, -1
```

```python
class ball_class():
    
    def __init__(self, gameDisplay):
        self.gameDisplay = gameDisplay
        
        self.ball_rect = BALL_RECT
        self.ball_radius = BALL_RADIUS
        
        #define the rectangle of the ball
        #Rect((left, top), (width, height))
        self.ball = pygame.Rect(DISPLAY_W//2, DISPLAY_H // 2, self.ball_rect, self.ball_rect)
        
        #moving speed
        self.dx = DX
        self.dy = DY
        self.ball_speed = BALL_SPEED
        self.paddle = paddle_class(gameDisplay)
        self.terminated = False

    def reset(self):
        self.terminated = False
        # when reset, the ball moves towards a random direction
        if np.random.rand() > 0.5:
            self.dx = self.dx
        else:
            self.dx = -self.dx
        if np.random.rand() > 0.5:
            self.dy = self.dy
        else:
            self.dy = -self.dy
        #set the ball at a random postion near the center
        self.ball = pygame.Rect(DISPLAY_W//2 + np.random.randint(-20,20), DISPLAY_H // 2 + np.random.randint(-20,20), self.ball_rect, self.ball_rect)
    
    def move(self):
        #if not hitting anything, keep moving
        self.ball.centerx += self.ball_speed * self.dx
        self.ball.centery += self.ball_speed * self.dy
        #collide on left and right
        if self.ball.centerx < self.ball_radius:
            self.dx = -self.dx
            self.ball.centerx = self.ball_radius + 1
        if self.ball.centerx > (DISPLAY_W - self.ball_radius):
            self.dx = -self.dx
            self.ball.centerx = (DISPLAY_W - self.ball_radius) - 1
            
        # collision top
        if self.ball.centery < self.ball_radius:
            self.dy = -self.dy
            self.ball.centery = self.ball_radius +1
            

    def check_collision(self):
        if self.ball.bottom > DISPLAY_H:
            self.terminated = True
            self.reset()

        #if hits the paddle, bounce
        if self.ball.colliderect(self.paddle.rect) and self.dy > 0:
            self.dx, self.dy = self.collision_move(self.paddle.rect)
        
    def collision_move(self,rect):
        #first, if the ball is moving towards right,
        #find out whether the ball is hitting from left side of a rect
        if self.dx > 0:
            delta_x = self.ball.right - rect.left
        # or the ball is moving left, see if it is hitting from the right side
        else:
            delta_x = rect.right - self.ball.left

        #check if the ball is hitting top or bottom
        if self.dy > 0:
            delta_y = self.ball.bottom - rect.top
        else:
            delta_y = rect.bottom - self.ball.top
    
        # the ball is hitting the corner, bounce back.
        if abs(delta_x - delta_y) < 10:
            self.dx, self.dy = -self.dx, -self.dy
        
        # the ball is hitting the top/bottom, only change dy
        elif delta_x > delta_y:
            self.dy = -self.dy
        
        # the ball is htting left/right side, only change dx
        elif delta_y > delta_x:
            self.dx = -self.dx

        return self.dx, self.dy
        
    def draw(self):
        #draw a cicle
        pygame.draw.circle(self.gameDisplay, pygame.Color('white'), self.ball.center, self.ball_radius)

    def update(self):
        self.draw()
        #update paddle in the loop
        self.paddle.update()
        self.move()
        self.check_collision()
```

```python
class blocks():

    def __init__(self, gameDisplay):
        self.gameDisplay = gameDisplay
        self.block_list = BLOCK_LIST
        self.color_list = COLOR_LIST

    def reset(self):
        self.block_list = [pygame.Rect(5 + 60 * i, 20 + 30 * j, 50, 15) for i in range(10) for j in range(4)]
        self.color_list = [(rnd(30, 256), rnd(30, 256), rnd(30, 256)) for i in range(10) for j in range(4)]
        [pygame.draw.rect(self.gameDisplay, self.color_list[color], block) for color, block in enumerate(self.block_list)]

    def update(self):
        [pygame.draw.rect(self.gameDisplay, self.color_list[color], block) for color, block in enumerate(self.block_list)]
```

```python
class ball_class():
    
    def __init__(self, gameDisplay):
        self.gameDisplay = gameDisplay
        
        self.ball_rect = BALL_RECT
        self.ball_radius = BALL_RADIUS
        
        self.ball = pygame.Rect(DISPLAY_W//2, DISPLAY_H // 2, self.ball_rect, self.ball_rect)
        
        self.dx = DX
        self.dy = DY
        self.ball_speed = BALL_SPEED
        
        self.paddle = paddle_class(gameDisplay)
        self.blocks = blocks(gameDisplay)
        self.blocks.reset()
        self.terminated = False
        

    def reset(self):
        self.terminated = False
        
        if np.random.rand() > 0.5:
            self.dx = self.dx
        else:
            self.dx = -self.dx
        if np.random.rand() > 0.5:
            self.dy = self.dy
        else:
            self.dy = -self.dy
        self.blocks.reset()
        self.ball = pygame.Rect(DISPLAY_W//2 + np.random.randint(-20,20), DISPLAY_H // 2 + np.random.randint(-20,20), self.ball_rect, self.ball_rect)
        self.paddle.set_position(PADDLE_START_X, PADDLE_START_Y)
    
    def move(self):
        #if not hitting anything, keep moving
        self.ball.centerx += self.ball_speed * self.dx
        self.ball.centery += self.ball_speed * self.dy
        #collide on left and right
        if self.ball.centerx < self.ball_radius:
            self.dx = -self.dx
            self.ball.centerx = self.ball_radius + 1
        if self.ball.centerx > (DISPLAY_W - self.ball_radius):
            self.dx = -self.dx
            self.ball.centerx = (DISPLAY_W - self.ball_radius) - 1
            
        # collision top
        if self.ball.centery < self.ball_radius:
            self.dy = -self.dy
            self.ball.centery = self.ball_radius +1
            
    def check_collision(self):
        if self.ball.bottom > DISPLAY_H:
            self.terminated = True
            self.reset()
            
        if self.ball.colliderect(self.paddle.rect) and self.dy > 0:
            self.dx, self.dy = self.collision_move(self.paddle.rect)

        #check the ball hits any blocks
        self.hit_index = self.ball.collidelist(self.blocks.block_list)

        # if there are still block remaining, pop that block from list
        if self.hit_index != -1:
            self.hit_rect = self.blocks.block_list.pop(self.hit_index)
            self.hit_color = self.blocks.color_list.pop(self.hit_index)
            self.dx, self.dy = self.collision_move(self.hit_rect)
            # special effect
            self.hit_rect.inflate_ip(self.ball.width * 3, self.ball.height * 3)
            pygame.draw.rect(self.gameDisplay, self.hit_color, self.hit_rect)
            
        if not len(self.blocks.block_list):
            for i in range(5):
                green = (0, 255, 0)
                blue = (0, 0, 128)
                font = pygame.font.Font('freesansbold.ttf', 32)
                text = font.render('You Win', True, green, blue)
                textRect = text.get_rect()
                textRect.center = (DISPLAY_W // 2, DISPLAY_H // 2)
                self.gameDisplay.blit(text, textRect)
            self.reset()            
        
    def collision_move(self,rect):
        # we need to find out which side the ball hits
        if self.dx > 0:
            delta_x = self.ball.right - rect.left
        else:
            delta_x = rect.right - self.ball.left
        if self.dy > 0:
            delta_y = self.ball.bottom - rect.top
        else:
            delta_y = rect.bottom - self.ball.top
    
        if abs(delta_x - delta_y) < 10:
            self.dx, self.dy = -self.dx, -self.dy
        elif delta_x > delta_y:
            self.dy = -self.dy
        elif delta_y > delta_x:
            self.dx = -self.dx

        return self.dx, self.dy
        
    def draw(self):
        pygame.draw.circle(self.gameDisplay, pygame.Color('white'), self.ball.center, self.ball_radius)

    def update(self):
        self.draw()
        self.paddle.update()
        self.blocks.update()
        self.move()
        self.check_collision()
```

```python
# add labels to the display
def update_label(data, title, font, x, y, gameDisplay):
    #first title, then data, '1' means anti-aliasing
    label = font.render('{}{}'.format(title,data),1,DATA_FONT_COLOR)
    # draw this label at x, y position
    # gameDisplay.blit means draw in the window
    gameDisplay.blit(label,(x,y))
    return y 
```

```python
class NeuralNetwork():
    def __init__(self, inputnodes, hiddennodes1, hiddennodes2, outputnodes):
        self.inodes=inputnodes
        self.hnodes1=hiddennodes1
        self.hnodes2=hiddennodes2
        self.onodes=outputnodes
        self.nnet_op = NNT_operation()
        self.epsilon = EPSILON

    def q_nnet_reset(self):
        self.wih = np.random.normal(0.0, pow(self.inodes, -0.5),(self.hnodes1, self.inodes))
        self.whh = np.random.normal(0.0, pow(self.hnodes1, -0.5),(self.hnodes2, self.hnodes1))
        self.who = np.random.normal(0.0, pow(self.hnodes2, -0.5),(self.onodes, self.hnodes2))
        #bias of hidden layer and output layer
        self.b_h1 = np.random.normal(0.0, pow(self.hnodes1, -0.5),(self.hnodes1))
        self.b_h2 = np.random.normal(0.0, pow(self.hnodes2, -0.5),(self.hnodes2))
        self.b_o = np.random.normal(0.0, pow(self.onodes, -0.5),(self.onodes))

    def action_selection(self, observation):
        
        random_number = np.random.random()
        
        #exploitation (choose the action that maximize Q)
        if random_number > self.epsilon:
            #take in four actions for a state
            nnet_predict = self.nnet_op.nnet_compute(observation, self.wih, self.whh, self.who, self.b_h1, self.b_h2, self.b_o)
            
            # find the action with max Q
            action = np.argmax(nnet_predict)
        
        #exploration (choose a random action)
        else:
            if np.random.randint(0,2) == 1:
                action = 1
            else:
                action = 0
        return action

    def reduce_epsilon(self,iteration):
        self.epsilon = self.epsilon * 0.998
        return self.epsilon
```

```python
class NNT_operation():

    def __init__(self):
        self.lr = LEARNING_RATE
        self.activation_function =  lambda x:scipy.special.expit(x)

    def nnet_compute(self, inputs, wih, whh, who, b_h1, b_h2, b_o):

        self.hidden_inputs1 = np.dot(wih, inputs) + b_h1
        self.hidden_outputs1 = self.activation_function(self.hidden_inputs1) 
        self.hidden_inputs2 = np.dot(whh, self.hidden_outputs1) + b_h2
        self.hidden_outputs2 = self.activation_function(self.hidden_inputs2)
        self.final_inputs = np.dot(who, self.hidden_outputs2) + b_o
        self.final_outputs = self.activation_function(self.final_inputs)
        print(self.final_outputs)
        
        return self.final_outputs
```

```python
# add labels to the display
def update_label(data, title, font, x, y, gameDisplay):
    #first title, then data, '1' means anti-aliasing
    label = font.render('{}{}'.format(title,data),1,DATA_FONT_COLOR)
    # draw this label at x, y position
    # gameDisplay.blit means draw in the window
    gameDisplay.blit(label,(x,y))
    return y 
```

```python
def update_data_labels(gameDisplay, game_time, num_iterations, reward, epsilon, font):
    y_pos = 150
    gap = 20
    x_pos = 10
    #update the y postion to 30 (return y)
    y_pos = update_label(round(game_time/60,2),'Game time', font, x_pos, y_pos + gap, gameDisplay)
    y_pos = update_label(num_iterations,'Iteration', font, x_pos, y_pos + gap, gameDisplay)
    y_pos = update_label(reward,'reward', font, x_pos, y_pos + gap, gameDisplay)
    y_pos = update_label(epsilon,'epsilon', font, x_pos, y_pos + gap, gameDisplay)
```

```python
class paddle_class():

    def __init__(self, gameDisplay):
        self.gameDisplay = gameDisplay
        self.img = pygame.image.load(PADDLE_IMG)
        self.rect = self.img.get_rect()
        self.set_position(PADDLE_START_X, PADDLE_START_Y)

    def set_position(self, x, y):
        self.rect.centerx = x
        self.rect.centery = y
        
    def draw(self):
        self.gameDisplay.blit(self.img, self.rect)

    def update(self):
        self.draw()
```

```python
class blocks():

    def __init__(self, gameDisplay):
        self.gameDisplay = gameDisplay
        self.block_list = block_list
        self.color_list = color_list

    def reset(self):
        self.block_list = [pygame.Rect(5 + 60 * i, 20 + 30 * j, 50, 15) for i in range(10) for j in range(4)]
        self.color_list = [(rnd(30, 256), rnd(30, 256), rnd(30, 256)) for i in range(10) for j in range(4)]
        [pygame.draw.rect(self.gameDisplay, self.color_list[color], block) for color, block in enumerate(self.block_list)]

    def update(self):
        [pygame.draw.rect(self.gameDisplay, self.color_list[color], block) for color, block in enumerate(self.block_list)]
```

```python
class ball_class():
    
    def __init__(self, gameDisplay):
        self.gameDisplay = gameDisplay
        self.img = pygame.image.load(PADDLE_IMG)
        self.ball_rect = ball_rect
        self.ball = pygame.Rect(DISPLAY_W//2, DISPLAY_H // 2, ball_rect, self.ball_rect)
        self.ball_radius = ball_radius
        self.dx = dx
        self.dy = dy
        self.ball_speed = ball_speed
        self.paddle = paddle_class(gameDisplay)
        self.blocks = blocks(gameDisplay)
        self.terminated = False
        self.iteration = 0
        self.reward = 0
        self.nnet = NeuralNetwork(INPUT_NODES, HIDDEN_NODES1, HIDDEN_NODES2, OUTPUT_NODES)
        self.nnet.q_nnet_reset()
                
        # now we initiate the weights of neural network
        self.nnet.target_nnet_update()
        
    def reset(self):
        self.terminated = False
        self.iteration +=1
        self.reward = 0
        if np.random.rand() > 0.5:
            self.dx = self.dx
        else:
            self.dx = -self.dx
        if np.random.rand() > 0.5:
            self.dy = self.dy
        else:
            self.dy = -self.dy
        self.blocks.reset()
        self.ball = pygame.Rect(DISPLAY_W//2 + np.random.randint(-20,20), DISPLAY_H // 2 + np.random.randint(-20,20), self.ball_rect, self.ball_rect)
        self.paddle.set_position(PADDLE_START_X, PADDLE_START_Y)
        self.ep = self.nnet.reduce_epsilon(self.iteration)
    
    def check_iteration_epsilon_reward(self):
        return self.iteration, self.ep, self.reward
    
    def move(self):
        self.ball.centerx += self.ball_speed * self.dx
        self.ball.centery += self.ball_speed * self.dy
        
        #collide on left and right
        if self.ball.centerx < self.ball_radius:
            self.dx = -self.dx
            self.ball.centerx = self.ball_radius + 1
        if self.ball.centerx > (DISPLAY_W - self.ball_radius):
            self.dx = -self.dx
            self.ball.centerx = (DISPLAY_W - self.ball_radius) - 1
            
        # collision top
        if self.ball.centery < self.ball_radius:
            self.dy = -self.dy
            self.ball.centery = self.ball_radius +1
            
    def check_collision(self):
        if self.ball.bottom > DISPLAY_H:
            self.terminated = True
            self.reward = -1            
            
        if self.ball.colliderect(self.paddle.rect) and self.dy > 0:
            self.dx, self.dy = self.collision_move(self.paddle.rect)
            self.reward = 1 

        self.hit_index = self.ball.collidelist(self.blocks.block_list)

        if self.hit_index != -1:
            self.hit_rect = self.blocks.block_list.pop(self.hit_index)
            self.hit_color = self.blocks.color_list.pop(self.hit_index)
            self.dx, self.dy = self.collision_move(self.hit_rect)
            # special effect
            self.hit_rect.inflate_ip(self.ball.width * 3, self.ball.height * 3)
            pygame.draw.rect(self.gameDisplay, self.hit_color, self.hit_rect)

        if not len(self.blocks.block_list):
            green = (0, 255, 0)
            blue = (0, 0, 128)
            font = pygame.font.Font('freesansbold.ttf', 32)
            text = font.render('You Win', True, green, blue)
            textRect = text.get_rect()
            textRect.center = (DISPLAY_W // 2, DISPLAY_H // 2)
            self.gameDisplay.blit(text, textRect)
            self.reset()

    def get_observation(self):
        # normalize the observations       
        obs1 = self.ball.centerx/600
        obs2 = self.ball.centery/400
        obs3 = self.dx*0.8
        obs4 = self.dy*0.8
        obs5 = self.paddle.rect.centerx/600
            
        observation = [obs1, obs2, obs3, obs4, obs5]
        return observation

    def take_action(self):
        self.old_state = self.get_observation()
        self.action = self.nnet.action_selection(self.old_state)    
        if self.action ==0 and self.paddle.rect.left > 0:
            self.paddle.rect.centerx -= PADDLE_SPEED
        if self.action ==1 and self.paddle.rect.right < DISPLAY_W:
            self.paddle.rect.centerx += PADDLE_SPEED 
        
    def collision_move(self,rect):
        if self.dx > 0:
            delta_x = self.ball.right - rect.left
        else:
            delta_x = rect.right - self.ball.left
        if self.dy > 0:
            delta_y = self.ball.bottom - rect.top
        else:
            delta_y = rect.bottom - self.ball.top
        if abs(delta_x - delta_y) < 10:
            self.dx, self.dy = -self.dx, -self.dy
        elif delta_x > delta_y:
            self.dy = -self.dy
        elif delta_y > delta_x:
            self.dx = -self.dx
        return self.dx, self.dy
        
    def draw(self):
        pygame.draw.circle(self.gameDisplay, pygame.Color('white'), self.ball.center, self.ball_radius)

    def pack_everything(self):
        self.new_state = self.get_observation()
        self.data_pack = self.old_state, self.action, self.reward, self.new_state, self.terminated
        
        #send the data to replay buffer
        self.nnet.replay_buffer(self.data_pack)
        
    def update(self):
        self.draw()
        self.take_action()
        self.move()
        self.paddle.update()
        self.blocks.update()
        self.check_collision()
        self.pack_everything()
        # train the network
        self.nnet.replay_and_learn()
        #reset reward to zero
        if self.reward == 1:
            self.reward =0
        if self.terminated == True:
            self.reset()        
```

```python
class NeuralNetwork():
    def __init__(self, inputnodes, hiddennodes1, hiddennodes2, outputnodes):
        self.inodes=inputnodes
        self.hnodes1=hiddennodes1
        self.hnodes2=hiddennodes2
        self.onodes=outputnodes
        self.replay = REPLAY
        self.nnet_op = NNT_operation()
        self.epsilon = EPSILON
        self.batch_size = BATCH_SIZE
        self.gamma = GAMMA

    def replay_buffer(self,datapack):
        self.replay.append(datapack)

    def q_nnet_reset(self):
        self.wih = np.random.normal(0.0, pow(self.inodes, -0.5),(self.hnodes1, self.inodes))
        self.whh = np.random.normal(0.0, pow(self.hnodes1, -0.5),(self.hnodes2, self.hnodes1))
        self.who = np.random.normal(0.0, pow(self.hnodes2, -0.5),(self.onodes, self.hnodes2))
        self.b_h1 = np.random.normal(0.0, pow(self.hnodes1, -0.5),(self.hnodes1))
        self.b_h2 = np.random.normal(0.0, pow(self.hnodes2, -0.5),(self.hnodes2))
        self.b_o = np.random.normal(0.0, pow(self.onodes, -0.5),(self.onodes))
        
    def target_nnet_update(self):
        #self.t_wih = np.clip(self.wih, -30, 30)
        self.t_wih = self.wih
        self.t_whh = self.whh
        self.t_who = self.who
        self.t_b_h1 = self.b_h1
        self.t_b_h2 = self.b_h2
        self.t_b_o = self.b_o
        #print('max and min of t_wih', np.max(self.t_wih),np.min(self.t_wih))
        #print('max and min of t_whh', np.max(self.t_whh),np.min(self.t_whh))
        #print('max and min of t_who', np.max(self.t_who),np.min(self.t_who))

    def action_selection(self, observation):
        random_number = np.random.random()
        if random_number > self.epsilon:
            nnet_predict = self.nnet_op.nnet_compute(observation, self.wih, self.whh, self.who, self.b_h1, self.b_h2, self.b_o)
            action = np.argmax(nnet_predict)
        else:
            if np.random.randint(0,2) == 1:
                action = 1
            else:
                action = 0
        return action

    def reduce_epsilon(self,iteration):
        self.epsilon = self.epsilon * 0.995
        return self.epsilon

    def epsilon_reset(self):
        self.epsilon = 1
        
    def replay_and_learn(self):

        #do not train the network when the replay buffer does not have enough data
        if len(self.replay) < self.batch_size:
            return
        #randomly sample 64 data from the replay buffer
        samples = random.sample(self.replay, self.batch_size)
        self.q_value = []
        self.target_value = []
        target_batch = []
        
        zipped_samples = list(zip(*samples))
        
        states,actions,rewards,new_states,terminated = zipped_samples

        for i in range(self.batch_size):
            # first step is to get all the values from the Q-network for state and action for time t: St, At
            q_value = self.nnet_op.nnet_compute(states[i], self.wih, self.whh, self.who, self.b_h1, self.b_h2, self.b_o)
            self.q_value.append(q_value)
        
        for i in range(self.batch_size):
            # second step is to get all the values from the target-network for St+1 and At+1
            target_value = self.nnet_op.nnet_compute(new_states[i], self.t_wih, self.t_whh, self.t_who, self.t_b_h1, self.t_b_h2, self.t_b_o)
            self.target_value.append(target_value)

        for j in range(self.batch_size):
            #find out the max Q value in the next state
            max_q_value = max(self.target_value[j])
                        
            # just make sure that we are not overwriting anything
            target = self.q_value[j].copy()
          
            if terminated[j]:
                target[actions[j]] = rewards[j]
            else:
                target[actions[j]] = rewards[j] + max_q_value*GAMMA
            
            target_batch.append(target)

        #train the q_network
        self.wih, self.whh, self.who, self.b_h1, self.b_h2, self.b_o = self.nnet_op.train(self.batch_size, self.wih, self.whh, self.who, self.b_h1, self.b_h2, self.b_o, states, target_batch)

    #def q_nnnet_train(self, input, target):
        #self.wih, self.whh, self.who = self.nnet_op.train(self.wih, self.whh, self.who, input, target)
    
```

```python
class NNT_operation():

    def __init__(self):
        self.lr = LEARNING_RATE
        self.activation_function =  lambda x:scipy.special.expit(x)

    def train(self, batch_size, wih, whh, who, b_h1, b_h2, b_o, inputs_list, targets_list):
        for i in range(batch_size):

            hidden_inputs1 = np.dot(wih, np.array(inputs_list[i])) + b_h1
            hidden_outputs1 = self.activation_function(hidden_inputs1)
            hidden_inputs2 = np.dot(whh, hidden_outputs1) + b_h2
            hidden_outputs2 = self.activation_function(hidden_inputs2)
            final_inputs = np.dot(who, hidden_outputs2) + b_o
            final_outputs = self.activation_function(final_inputs)
                        
            output_errors = np.array(targets_list[i]) - final_outputs
            hidden_errors2 =  np.dot(who.T, output_errors)
            hidden_errors1 =  np.dot(whh.T, hidden_errors2)

            who += self.lr*np.dot(np.transpose(np.array([output_errors*final_outputs*(1.0 - final_outputs)])),([hidden_outputs2]))
            b_o += self.lr*(output_errors * final_outputs*(1.0-final_outputs))
            whh += self.lr*np.dot(np.transpose(np.array([hidden_errors2*hidden_outputs2*(1.0 - hidden_outputs2)])),([hidden_outputs1]))
            b_h2 += self.lr*(hidden_errors2 * hidden_outputs2*(1.0-hidden_outputs2))
            wih += self.lr*np.dot(np.transpose(np.array([hidden_errors1*hidden_outputs1*(1.0 - hidden_outputs1)])),([inputs_list[i]]))
            b_h1 += self.lr*(hidden_errors1 * hidden_outputs1*(1.0-hidden_outputs1))

            #who = np.clip(who, -3, 3)
            #whh = np.clip(whh, -3, 3)
            #wih = np.clip(wih, -3, 3)
        
        return wih, whh, who, b_h1, b_h2, b_o

    def nnet_compute(self, inputs, wih, whh, who, b_h1, b_h2, b_o):
                
        self.hidden_inputs1 = np.matmul(wih, inputs) + b_h1
        self.hidden_outputs1 = self.activation_function(self.hidden_inputs1) 
        self.hidden_inputs2 = np.matmul(whh, self.hidden_outputs1) + b_h2
        self.hidden_outputs2 = self.activation_function(self.hidden_inputs2)
        self.final_inputs = np.matmul(who, self.hidden_outputs2) + b_o
        self.final_outputs = self.activation_function(self.final_inputs)
        
        return self.final_outputs
```

```python
def run_game():
    pygame.init()
    gameDisplay = pygame.display.set_mode((DISPLAY_W, DISPLAY_H))
    pygame.display.set_caption('Break Out Machine Learning Example')
    label_font = pygame.font.SysFont("monospace", DATA_FONT_SIZE)
    
    
    bgimg = pygame.image.load(BG_IMG)
    ball = ball_class(gameDisplay)
    ball.reset()

    running =  True
    clock = pygame.time.Clock()

    dt = 0
    game_time = 0

    num_iterations = 1

    step = 0

    while running:

        step +=1

        #update the target net every 100 steps
        if step % 1000 == 0:
            ball.nnet.target_nnet_update()
            
        dt = 1
        game_time += dt
        gameDisplay.blit(bgimg, (0, 0))

        ball.update()

        num_iterations, epsilon, reward = ball.check_iteration_epsilon_reward()
        
        update_data_labels(gameDisplay, game_time, num_iterations, reward, epsilon, label_font)

        pygame.display.update() 
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
                pygame.display.quit() 
```
